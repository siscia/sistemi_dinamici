### Punto 1

Gli assi positivi del sistema sono invarianti, infatti per $x_1 = 0$ si ha $\dot x_1 = 0$ e per $x_2 = 0$ similmente di ottiene $\dot x_2 = 0$. Questo significa che un qualunque punto nel primo quadrante rimarrà "intrappolato" dato che non potrà mai "sfondare" gli assi visto la derivata nulla sugli assi stessi.

### Punto 2

Userò l'asse orizzontale per la variabile $x_1$ e quello verticale per la variabile $x_2$.

Potendo scegliere sembrerebbè più conveniente fare lo studio per $x_2(0) = 0$ che porta a una equazioni matematiche più semplice, una semplice parabola.

Osservando il sistema con più attenzione e capendo cosa il sistema sta modelizzando però si nota che ponendo $x_1(0) = 0$ si ottiene un equazione di $\dot x_2$ matematicamente più complessa ma sempre negativa, infatti: $$\dot x_2 = -mx_2 - \frac{2px_2}{1+2x_2} = -(mx_2+ \frac{2px_2}{1+2x_2})$$ con sia $m$ e $p$ sempre positivi per ipotesi.

L'equazione ha due soluzioni, la soluzione banale $x_2 = 0$ ed una seconda soluzione non banale dipendente dal parametro $p$ negativa e quindi non tratto dato che non avrebbe senso nello studio di questo sistema reale.

Lo studio ci porta a dire che il numero di consumatori calerà fino all'equilibrio sul punto $(0,0)$, non ci sono altri punti di equilibrio o di particolare interesse nel sistema reale.

### Punto 2 bis

Userò l'asse orizzontale per la variabile $x_1$ e quello verticale per la variabile $x_2$.

Potendo scegliere preferisco fare lo studio per $x_2(0) = 0$ questo perchè già sappiamo che $\dot x_2 = 0$ e inoltre vedo velocemente che $$\dot x_1 = r x_1 (1 - x_1) = - r x_1^2 + r x_1 $$ è una semplice parabola.

L'inverso $x_1(0) = 0$ avrebbe generato un $\dot x_2$ leggermente più complesso.

Come sappiamo gli assi sono invarianti del sistema, un punto che parte su un asse rimarrà su quell'asse per sempre, il nostro punto quindi rimarrà per sempre sull'asse orizontale $x_2 = 0$, varierà però la sua componente orizontale $x_1$.

Cerchiamo il punto di equilibrio sull'asse orizontale $$\dot x_1 = 0 = -rx_1^2+r_x1 = rx_1(1-x_1)$$ da cui otteniamo due soluzioni non coincidenti. 

La soluzione banale $$ x_1 = 0 $$ il che non ci sorprende assolutamente dato che lo avevamo già capito nel punto precendete, il che significa anche che il punto $(0,0)$ è un equilibrio del sistema, riflettendo su cosa il modello rappresenta è ovvio che in assenza di prede (risorse) e di predatori (consumatori) il loro numerò non può che rimanere zero.

L'altra soluzione $$x_1 = 1$$ che porta il numero di risorse ad una quantità constante, la capacità portante del sistema.

Il punto di equilibrio $x_1 = 1$ è punto di equilibrio stabile, infatti per $x_1 = 1 + \delta $ (risorse sopra la capacità portante) si ottiene $ \dot x_1 = rx_1(- \delta) < 0 $ che significa che le risorse diminuiscono e per $x_1 = 1 - \delta $ (risorse sotto la capacità portante) si ottiene similmente $ \dot x_1 = rx_1\delta > 0$ che significa che le risorse stanno aumentando quindi il sistema tende verso $x_1 = 1$

```python
r = Par(101.0/100, 'r')
m = Par(0.122, 'm')
p = Par(0.65, 'p')
params = [pa / 100.0 for pa in xrange(51, 75, 3)]
x1 = Var('x1')
x2 = Var('x2')
x1_dot = r*x1*(1-x1) - (5*x1*x2)/(1+3*x1)
x2_dot = (5*x1*x2)/(1+3*x1) - m*x2 - (2*p*x2)/(1+2*x2)
variable = {x1 : x1_dot, x2 : x2_dot}
initial_condition = {x1 : 0.4, x2 : 0.2}
```

```python
DSargs = args()
DSargs.name = 'RisorsaConsumatore'
DSargs.ics = initial_condition
DSargs.pars = [r, m, p]
DSargs.tdata = [0, 50]
DSargs.varspecs = variable
DSargs.pdomain = {'p' : [0.51, 0.75]}
```

```python
DS = Generator.Vode_ODEsystem(DSargs)
```

```python
for px in params:
    DS.set(pars={'p' : px})
    traj = DS.compute('Preda-Consumatore p = ' + str(px))
    pts = traj.sample()
    plt.figure()
    plt.plot(pts['t'], pts['x1'], label = 'Prede')
    plt.plot(pts['t'], pts['x2'], label = 'Consumatori')
    plt.legend()
    plt.xlabel('t')
```

```python
x1_dot = Var('r*x1*(1-x1) - (5*x1*x2)/(1+3*x1)', 'v')
x1_dot_fun = expr2fun(x1_dot, r = r.eval())
x2_dot = Var('(5*x1*x2)/(1+3*x1) - m*x2 - (2*p*x2)/(1+2*x2)', 'v')
x2_dot_fun = expr2fun(x2_dot, m = m.eval(), p = p.eval())
x2_dot_fun_param = expr2fun(x2_dot, m = m.eval())
```

```python
plt.figure(1)
x1 = np.arange(0, 1, 0.001)
x2 = np.arange(0, 1, 0.001)
X1, X2 = meshgrid(x1, x2)

plt.contour(X1, X2, x1_dot_fun(X1, X2), [0])

for px in params:
    plt.contour(X1, X2, x2_dot_fun_param(X1, X2, px), [0])
```

L'isoclina di $$ \dot{x_1} = rx_1(1-x_1)-\frac{5x_1x_2}{1+3x_1} = 0 $$ è una parabola a meno di $\frac{5x_1x_2}{1+3x_1}$

Sotto la parabola disegnata da $ \dot x_1 = 0 $ abbiamo $ \dot x_1 > 0 $, al contrario sopra l'isoclina abbiamo $ \dot x_1 < 0 $ come già visto in 2.bis

L'isoclina di $$ \dot{x_2} = \frac{5x_1x_2}{1+3x_1} - mx_2 - \frac{2px_2}{1+2x_2} = 0$$ 

divide il piano in modo tale da ricordare l'iperbole $ xy = 1 $, il termine $-\frac{2 p x_2}{1+2x_2} $ alza o abbassa la curva, per opportuni valori quindi la curva può incrociare l'isoclina di $x_1$ in due punti distinti generando due equilibri positivi.

Per terminare l'analisi della seconda isoclina possiamo provare a metterci in un punto sotto isoclina di $x_2$ in una situazione con tanti consumatori e poche risorse, è ovvio che il numero dei consumatori calerà.

$$ \dot x_2(x_1 = 0.1, x_2 = 0.5, p = 0.75) \approx -0.19$$

Similmente se ci posizioniamo sopra l'iperbole in una situazione con molti consumatori e tante risorse il numero dei consumatori tenderà a crescere.

$$ \dot x_2(x_1 = 0.6, x_2 = 0.6, p = 0.5) \approx 0.30 $$

Il parametro $p$ potrebbe modellare la competizione interna ai consumatori, il termine $- \frac{2px_2}{1+2x_2}$ cresce infatti con l'aumentare dei consumatori il che porta ad una maggiore competizione per l'uso delle risorse aumentando l'attrito nel sistema e rendendo necessarie più risorse per i consumatori.

```python
for px in params:
    plt.figure()
    plt.streamplot(X1, X2, x1_dot_fun(X1,X2), x2_dot_fun_param(X1, X2, px))
    plt.contour(X1, X2, x2_dot_fun_param(X1, X2, px), [0])
    plt.contour(X1, X2, x1_dot_fun(X1, X2), [0])
```

Il campo vettoriale identifica la presenza di un ciclo nel primo equilibrio e di una sella nel secondo equilibrio.
