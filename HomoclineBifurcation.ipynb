```python
from PyDSTool import *

r = Par(101.0/100, 'r')
m = Par(0.122, 'm')
p = Par(0.65, 'p')
params = [pa / 100.0 for pa in xrange(51, 75, 3)]
x1 = Var('x1')
x2 = Var('x2')
x1_dot = r*x1*(1-x1) - (5*x1*x2)/(1+3*x1)
x2_dot = (5*x1*x2)/(1+3*x1) - m*x2 - (2*p*x2)/(1+2*x2)

variable = {x1 : x1_dot, x2 : x2_dot}
initial_condition = {x1 : 0.4, x2 : 0.2}

point = {'x1' : 0.567266396673,
         'x2' : 0.236170178391}

DSargs = args()
DSargs.name = 'RisorsaConsumatore'
DSargs.ics = initial_condition
DSargs.pars = [r, m, p]
DSargs.tdata = [0, 50]
DSargs.varspecs = variable
DSargs.pdomain = {'p' : [0.49, 0.75]}
```

```python
Fast = Generator.Dopri_ODEsystem(DSargs)
```

## First Bifurcation 

```python
p = 0.584704589844 

Fast.set(ics = {'x1' : point['x1'], 'x2' : point['x2']},
             tdata = [0, 3000],
             pars = {'p' : p})
print p
traj = Fast.compute('Preda-Consumatore p = ' + str(p))
pts = traj.sample()
plt.figure()
plt.plot(pts['x2'], pts['x1'], label = str(p))
plt.legend()
plt.xlabel('x2')
plt.ylabel('x1')
```

```python
p = (0.584704589844 + 0.584711914063) / 2
Fast.set(ics = {'x1' : point['x1'], 'x2' : point['x2']},
             tdata = [0, 3000],
             pars = {'p' : p})
print p
traj = Fast.compute('Preda-Consumatore p = ' + str(p))
pts = traj.sample()
plt.figure()
plt.plot(pts['x2'], pts['x1'], label = str(p))
plt.legend()
plt.xlabel('x2')
plt.ylabel('x1')
```
